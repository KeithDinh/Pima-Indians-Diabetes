### 7.	Fit a linear model that predicts the class attribute using the 8 z-scored, continuous attributes of the cleaned dataset as independent variables . Report the R2 of the linear model and the coefficients of each attribute in the obtained regression function.  Next, drop the two attributes, whose coefficients are the closest to 0, and obtain a linear model using the six remaining attributes as independent variable.  Do the coefficients tell you anything about the importance of the attribute in predicting diabetes? What about negative and positive coefficients? Also compare the two regression functions!

```python
df = diabetes_clean
cols = list(df.columns)
cols.remove('Outcome')
for z in cols:
    col_zscore = z + '_zscore'
    df[col_zscore] = (df[z] - df[z].mean())/df[z].std(ddof=0)
df

import statsmodels.formula.api as sm
result = sm.ols(formula="Outcome ~ Pregnancies_zscore + Glucose_zscore + BloodPressure_zscore + SkinThickness_zscore + Insulin_zscore + BMI_zscore + DiabetesPedigreeFunction_zscore + Age_zscore", data=df).fit()
print(result.params)
print(result.summary())
```

#### Drop SkinThickness_zscore and Insulin_zscore as two closest to 0

```python
result_after_drop = sm.ols(formula="Outcome ~ Pregnancies_zscore + Glucose_zscore + BloodPressure_zscore + BMI_zscore + DiabetesPedigreeFunction_zscore + Age_zscore", data=df).fit()
print(result_after_drop.params)
print(result_after_drop.summary())
```

After dropping the two columns, the statistics does not have any significant change. This is true because the coeffictions of the two columns are very small that do not leave any impact to the overall model.
Also, from the list of coeffictions, BloodPressure_zscore and Insulin_zscore are the only two having negative relationship with outcome. Among all, Glucose is having the strongest impact which means people with diabetes more likely to have high Glucose.
On the other hand, SkinThickness has the lowest coeffiction which is not considered to predict diabetes.

```python
pima = pd.read_csv('diabetes_cleaned.csv',names = ['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age','Outcome'])
#getting rid of the first row (header) : pregnancies, glucose, bloodpressure, skinthickness, ... 
pima = pima.iloc[1:]
#split dataset in features and target variable
feature_cols = ['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']
X = pima[feature_cols] 
y = pima.Outcome 
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)


##optimizing performance of decision tree
# Create Decision Tree classifer object
clf = DecisionTreeClassifier(criterion="entropy", max_depth=3)

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)

dot_data = StringIO()
export_graphviz(clf, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True, feature_names = feature_cols,class_names=['0','1'])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())

```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
